```python
import pandas as pd
import matplotlib.pyplot as plt

# Загрузка данных из CSV файла
file_path = "results.csv"
data = pd.read_csv(file_path)

# Печать загруженных данных
print(data)

# Построение графиков
plt.figure(figsize=(12, 6))

# График времени выполнения на CPU и GPU
plt.subplot(1, 2, 1)
plt.plot(data['n'], data['CPU_Time(s)'], label='CPU Time (s)', marker='o')
plt.plot(data['n'], data['GPU_Time(s)'], label='GPU Time (s)', marker='o')
plt.title('Время выполнения на CPU и GPU')
plt.xlabel('Размер матриц (n)')
plt.ylabel('Время (секунды)')
plt.legend()
plt.grid()

# График ускорения
plt.subplot(1, 2, 2)
plt.plot(data['n'], data['Speedup'], label='Speedup', marker='o', color='orange')
plt.title('Ускорение при использовании GPU')
plt.xlabel('Размер матриц (n)')
plt.ylabel('Ускорение')
plt.legend()
plt.grid()

# Показать графики
plt.tight_layout()
plt.show()
```

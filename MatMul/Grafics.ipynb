```python
import pandas as pd
import matplotlib.pyplot as plt

#    CSV 
file_path = "results.csv"
data = pd.read_csv(file_path)

#   
print(data)

#  
plt.figure(figsize=(12, 6))

#     CPU  GPU
plt.subplot(1, 2, 1)
plt.plot(data['n'], data['CPU_Time(s)'], label='CPU Time (s)', marker='o')
plt.plot(data['n'], data['GPU_Time(s)'], label='GPU Time (s)', marker='o')
plt.title('   CPU  GPU')
plt.xlabel('  (n)')
plt.ylabel(' ()')
plt.legend()
plt.grid()

#  
plt.subplot(1, 2, 2)
plt.plot(data['n'], data['Speedup'], label='Speedup', marker='o', color='orange')
plt.title('   GPU')
plt.xlabel('  (n)')
plt.ylabel('')
plt.legend()
plt.grid()

#  
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Загрузка данных из CSV файла
file_path = r"C:\CUDALAB\CUDA2\CUDA2\results.csv"
data = pd.read_csv(file_path)

# Удаление лишних пробелов в названиях столбцов
data.columns = data.columns.str.strip()

# Построение графиков
plt.figure(figsize=(12, 6))

# График времени выполнения на CPU (последовательное и многопоточное) и GPU
plt.subplot(1, 2, 1)
plt.plot(data['Vector Size'], data['CPU Time (s)'], label='Последовательное (s)', marker='o')
plt.plot(data['Vector Size'], data['CPU Multithreaded Time (s)'], label='CPU (2) (s)', marker='o')
plt.plot(data['Vector Size'], data['GPU Time (s)'], label='GPU (s)', marker='o')
plt.title('Время выполнения на CPU и GPU')
plt.xlabel('Размер вектора')
plt.ylabel('Время (секунды)')
plt.legend()
plt.grid()

# График ускорения
plt.subplot(1, 2, 2)
plt.plot(data['Vector Size'], data['Speedup CPU/Multithreaded'], label='Ускорение CPU(2)', marker='o', color='orange')
plt.plot(data['Vector Size'], data['Speedup CPU/GPU'], label='Ускорение GPU', marker='o', color='green')
plt.title('Ускорение CPU и GPU')
plt.xlabel('Размер вектора')
plt.ylabel('Ускорение')
plt.legend()
plt.grid()

# Показать графики
plt.tight_layout()
plt.show()
```

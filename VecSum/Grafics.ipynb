```python
import pandas as pd
import matplotlib.pyplot as plt

#    CSV 
file_path = r"C:\CUDALAB\CUDA2\CUDA2\results.csv"
data = pd.read_csv(file_path)

#      
data.columns = data.columns.str.strip()

#   
print(data)

#  
plt.figure(figsize=(14, 6))

#     CPU (  )  GPU (atomic  reduction)
plt.subplot(1, 2, 1)
plt.plot(data['Vector Size'], data['CPU Time (s)'], label=' (s)', marker='o')
plt.plot(data['Vector Size'], data['CPU Multithreaded Time (s)'], label='CPU (2) (s)', marker='o')
plt.plot(data['Vector Size'], data['GPU Atomic Time (s)'], label='GPU (Atomic) (s)', marker='o')
plt.plot(data['Vector Size'], data['GPU Reduction Time (s)'], label='GPU (Reduction) (s)', marker='o')
plt.title('   CPU  GPU')
plt.xlabel(' ')
plt.ylabel(' ()')
plt.legend()
plt.grid()

#  
plt.subplot(1, 2, 2)
plt.plot(data['Vector Size'], data['Speedup CPU/Multithreaded'], label=' CPU(2)', marker='o', color='orange')
plt.plot(data['Vector Size'], data['Speedup CPU/GPU Atomic'], label=' GPU (Atomic)', marker='o', color='green')
plt.plot(data['Vector Size'], data['Speedup CPU/GPU Reduction'], label=' GPU (Reduction)', marker='o', color='blue')
plt.title(' CPU  GPU')
plt.xlabel(' ')
plt.ylabel('')
plt.legend()
plt.grid()

#  
plt.tight_layout()
plt.show()
```
